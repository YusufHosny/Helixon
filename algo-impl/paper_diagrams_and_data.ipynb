# Spiral Model V.s. Ground Truth

```python
from model.spiral_model import Spiral
from hlxon_hdf5io import *
from matplotlib import pyplot as plt
import numpy as np
```

```python
X, Y, Z = 0, 1, 2
```

```python
# creating spiral model
spiral_pitch = 4.2 #m
spiral_radius = 8 #m
path_width = 2.4 #m
spiral = Spiral(spiral_pitch, spiral_radius, path_width)
```

```python
dataset = readAll()

fig = plt.figure()
ax = plt.axes(projection='3d')

for sequence in dataset:
    _, _, _, _, _, _, _, gt_timestamp, gt_position, gt_orientation = sequence
    spiral.align_to_spiral(gt_position)

    ax.plot3D(gt_position[:, X], gt_position[:, Y], gt_position[:, Z], color='gray', label='Ground Truth')
ax.plot3D(*spiral.point_at_z(dataset[3][-2][:, Z]), color='blue', label='Spiral Model')
ax.set_title('Spiral Model Vs Ground Truth')
```
